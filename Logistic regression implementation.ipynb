## Data pre-processing
### Features chosen
1. age:age in years <br />
2. sex-:(1 = male; 0 = female) <br />
3. cp:chest pain type <br />
4. trestbps-resting blood pressure (in mm Hg on admission to the hospital) <br />
5. chol:serum cholestoral in mg/dl <br />
6. fbs:(fasting blood sugar > 120 mg/dl) (1 = true; 0 = false) <br />
7. restecg:resting electrocardiographic results <br />
8. thalach:maximum heart rate achieved <br />
9. exang:exercise induced angina (1 = yes; 0 = no) <br />
10. oldpeak:ST depression induced by exercise relative to rest <br />
11. slope:the slope of the peak exercise ST segment <br />
12. ca:number of major vessels (0-3) colored by flourosopy <br />
13. thal:3 = normal; 6 = fixed defect; 7 = reversable defect <br />
14. target:1 or 0

```python
import pandas as pd

dataset = pd.read_csv("heart.csv")
dataset.head()
```

```python
# detailed statistical info of the dataset 
dataset.describe().transpose()
```

### Feature scaling

```python
dataset.columns
```

```python
# age is taken in groups as a categorical variable
# target variable is not normalized as 
cols_to_norm = ['sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalach','exang', 'oldpeak', 'slope', 'ca', 'thal']

# OR use the MinMaxScalar in sklearn
dataset[cols_to_norm] = dataset[cols_to_norm].apply(lambda x: (x - x.min()) / (x.max() - x.min()))

dataset.head()
```

### Logistic Regression Implementation

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
lr = LogisticRegression()
```

```python
dataset.columns.values
```

```python
cols = ['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalach',
       'exang', 'oldpeak', 'slope', 'ca', 'thal']
```

```python
X = dataset[cols]
y = dataset['target']
```

```python
lr.fit(X,y)
```

```python
lr.score(X,y)
```

```python
results = lr.predict_proba([X.iloc[1]])
print ("Chance of getting heart attack:", results[0][1]*100, '%')
```
